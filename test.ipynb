```python
import numpy as np
import matplotlib.pyplot as plt


a=1
b=0.5
c=0.02

list = []
for x in np.arange(0,200,1):
    value = (1 - np.exp(-c * x)) * (a - b) + b
    list.append(value)


plt.plot(np.arange(0,200,1), list)
plt.title('Line Plot of Computed Values')
plt.xlabel('Epoch')
plt.ylabel('Value')
plt.grid(True)
plt.show()
```
